**Question 6**

**check weather given string has vowel or consonants**

```python
def check_vowels_consonants(s):
    vowels = "aeiouAEIOU"
    has_vowel = False
    has_consonant = False

    for char in s:
        if char.isalpha():
            if char in vowels:
                has_vowel = True
            else:
                has_consonant = True

    if has_vowel and has_consonant:
        return "The string has both vowels and consonants."
    elif has_vowel:
        return "The string has only vowels."
    elif has_consonant:
        return "The string has only consonants."
    else:
        return "The string has no alphabetic characters."

string = input("Enter a string: ")
print(check_vowels_consonants(string))
```

**Question 10**

**check for weekday or weekend**

```python
def check_day(day):
    day = day.strip().lower()

    weekdays = ["monday", "tuesday", "wednesday", "thursday", "friday"]
    weekends = ["saturday", "sunday"]

    if day in weekdays:
        return f"{day.capitalize()} is a weekday."
    elif day in weekends:
        return f"{day.capitalize()} is a weekend."
    else:
        return "Invalid day entered."

day_input = input("Enter a day: ")
print(check_day(day_input))
```

```python
def grading_system(marks):
    if marks >= 90:
        return "Grade: A+"
    elif marks >= 80:
        return "Grade: A"
    elif marks >= 70:
        return "Grade: B"
    elif marks >= 60:
        return "Grade: C"
    elif marks >= 50:
        return "Grade: D"
    else:
        return "Grade: F (Fail)"

try:
    marks = int(input("Enter your marks (0-100): "))
    if 0 <= marks <= 100:
        print(grading_system(marks))
    else:
        print("Marks should be between 0 and 100.")
except ValueError:
    print("Please enter a valid number.")
```

**Question 13**

**check the type of triangle**

```python
def classify_triangle_by_sides(a, b, c):
    if a == b == c:
        return "Equilateral Triangle"
    elif a == b or b == c or a == c:
        return "Isosceles Triangle"
    else:
        return "Scalene Triangle"

def classify_triangle_by_angles(a, b, c):
    if a + b + c != 180 or a <= 0 or b <= 0 or c <= 0:
        return "Not a valid triangle"

    if a == 90 or b == 90 or c == 90:
        return "Right-angled Triangle"
    elif a > 90 or b > 90 or c > 90:
        return "Obtuse Triangle"
    else:
        return "Acute Triangle"

print("Classify by sides:")
side1 = float(input("Enter first side: "))
side2 = float(input("Enter second side: "))
side3 = float(input("Enter third side: "))
print(classify_triangle_by_sides(side1, side2, side3))

print("\nClassify by angles:")
angle1 = float(input("Enter first angle: "))
angle2 = float(input("Enter second angle: "))
angle3 = float(input("Enter third angle: "))
print(classify_triangle_by_angles(angle1, angle2, angle3))
```

**Question 15**

**check weather a number is a armstrong number or not**


```python
def is_armstrong(num):
    digits = str(num)
    power = len(digits)

    total = sum(int(digit) ** power for digit in digits)

    if total == num:
        return f"{num} is an Armstrong number."
    else:
        return f"{num} is not an Armstrong number."
try:
    number = int(input("Enter a number: "))
    print(is_armstrong(number))
except ValueError:
    print("Please.")
```

**Question 16**

**check weather the given string is palindrome**

```python
def is_palindrome(s):
    s = s.replace(" ", "").lower()

    if s == s[::-1]:
        return "The string is a palindrome"
    else:
        return "The string is not a palindrome"
string = input("Enter a string: ")
print(is_palindrome(string))
```
