*LOOPS*

```python
# print first 10 natural number using while loop
i = 1
while i <= 10:
    print(i)
    i += 1


```

```python
# print the pasacal's triangle pattern using for loop
row = 5
for i in range(1, row + 1):
    for j in range(1, i + 1):
        print(j, end = " ")
    print()


```

```python
# calculate the sum of all numbers from 1 to a given number using loops

sum = 0
x = int(input("enter no: "))
for i in range(1, x + 1):
    sum += i

print("the total is:",sum)
```

```python
# print multiplication table of a number using loops

x = int(input("enter no: "))
for i in range(1, 11, 1):
  mult = x*i
  print(mult)
```

```python
# display numbers from a list using a loop
# 1. The number must be divisible by 5
# 2. The number shoulbe be greater than 180
# 3. If the number is greater than 500 then stop the loop

Numbers = [12, 75, 150, 180, 145, 525, 50, 877]

for i in Numbers:
  if i > 150:
    if i > 500:
      break
    elif i % 5 == 0:
      print(i)


```

```python
# count the total number of digits in a number

num = int(input("enter no: "))
count = 0
while num != 0:
    num //= 10
    count += 1
print("total digits are:", count)

```

```python
# print the pattern

rows = 7
col = 7

for i in range(0, rows+1):
  for  j in range(col-i, 0, -1):

    print(j, end = " ")
  print()
```

```python
# print fibonacci series using loops

n = 10
a, b = 0, 1
count = 0

print("Fibonacci series:")
while count < n:
    print(a)
    a, b = b, a + b
    count += 1
```

```python
# find the factorial of a given number

num = 6
fac = 1

if num < 0:
  print("factorial does not exist for negative numbers")
elif num == 0:
  print("the factorial of 0 is 1")
else:
  for i in range(1, num +1):
    fac = fac * i
  print("the factorial of", num, "is", fac)
```

```python
# reverse a integer number
# given = 765432, expected output: 234567

num = 76432
rev = 0

print("given number is:", num)

while num > 0:
  rem = num % 10
  rev = (rev * 10) + rem
  num = num // 10

print("reversed number is:", rev)
```
