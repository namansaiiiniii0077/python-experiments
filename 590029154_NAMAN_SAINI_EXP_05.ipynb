```python
def find_max_min (numbers):
    nums = list(numbers)
    maximum = nums[0]
    minimum = nums[0]

    for num in nums:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num

    return maximum, minimum
x = {10, 20, 30, 40, 50}
max, min = find_max_min(x)
print("maximum:", max)
print("minimum:", min)
```

```python
# QS.2 Create a tuple to store n numeric values and find average of all values

n = int(input("enter values to input : "))
values_list = []
for i in range(n):
    val = float(input(f"enter value {i+1}: "))
    values_list.append(val)
values_tuple = tuple(values_list)
total = 0
for v in values_tuple:
    total += v

average = total / n
print("tuple of values:", values_tuple)
print("average of values:", average)
```
