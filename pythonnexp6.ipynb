```python
# write a python progarm to find the maximum and minimum numbers from a sequence of numbers (do not use built in functions)

def find_extremes(numbers):
    max = numbers[0]
    min = numbers[0]
    for n in numbers:
        if n > max:
            max = n
        if n < min:
            min = n

    return max, min
mylist = [2, 6, 4, 15, 16, 19, 99]
highest, lowest = find_extremes(mylist)

print("Highest:", highest)
print("Lowest:", lowest)
```

```python
#  Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.

def sum_of_cubes(n):
  total_sum = 0
  for i in range(1, n):
    total_sum += i**3
  return total_sum

n = int(input("enter a positive integer: "))
number = n
result = sum_of_cubes(n)
print(f"The sum of cubes smaller than {number} is: {result}")
```

```python
# Write a recursive function to print Fibonacci series upto n terms

def fibonacci(n):
  if n <= 1:
    return n
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)
terms = int(input("enter the number of terms: "))

print(f"fibonacci series up to {terms} terms:")
for i in range(terms):
    print(fibonacci(i), end=" ")
```

```python
# Write a lambda function to find volume of cone

import math
conevolume = lambda r, h: (1/3) * math.pi * r * r * h
print(conevolume(5, 7))
```

```python
#Write a lambda function which gives tuple of max and min from a list. Sample input: [10, 6, 8, 90, 12, 56] Sample output: (90,6)

get_max_min = lambda data: (max(data), min(data))
numbers = [10, 6, 8, 90, 12, 56]
result = get_max_min(numbers)
print(result)
```

```python
#Write functions to explain the mentioned concepts: a. Keyword argument
#b. Default argument
#c. Variable- length argument

def describe_person(name, age):
    print(f"{name} is {age} years old.")

# Using keyword arguments
describe_person(age=30, name="Alice")
def greet(name, message="Hello"):
    print(f"{message}, {name}!")

# Uses default message
greet("Bob")
greet("Charlie", message="Good morning")
def sum_numbers(*args):
    return sum(args)

print(sum_numbers(1, 2, 3))
print(sum_numbers(10, 20, 30, 40))
```

```python
#Write a program to check whether all the values in a dictionary are the same or not using a lambda function
def check_all_values_same(data_dict):

    if not data_dict:
        return True
    values = list(data_dict.values())
    check_same = lambda vals: all(item == vals[0] for item in vals)

    return check_same(values)
dict1 = {'a': 10, 'b': 10, 'c': 10}
dict2 = {'a': 10, 'b': 20, 'c': 10}

print(f"Dictionary 1 is uniform: {check_all_values_same(dict1)}")
print(f"Dictionary 2 is uniform: {check_all_values_same(dict2)}")
```
