```python
#Find a factorial of given number
def calculate_factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."

    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    return factorial
num = 5
result = calculate_factorial(num)

print(f"The factorial of {num} is: {result}")
```

```python
#Find whether the given number is Armstrong number.
def is_armstrong(number):
    num_str = str(number)
    num_digits = len(num_str)
    sum_of_powers = 0
    for digit in num_str:
        sum_of_powers += int(digit) ** num_digits
    return sum_of_powers == number
num = 153
if is_armstrong(num):
    print(f"{num} is an Armstrong number.")
else:
    print(f"{num} is not an Armstrong number.")
```

```python
#Print Fibonacci series up to given term.
def print_fibonacci(terms):
    a, b = 0, 1
    count = 0
    if terms <= 0:
        print("Please enter a positive integer.")
    elif terms == 1:
        print(f"Fibonacci sequence up to {terms} term:")
        print(a)
    else:
        print(f"Fibonacci sequence up to {terms} terms:")
        while count < terms:
            print(a, end=" ")
            nth = a + b
            a = b
            b = nth
            count += 1
        print()
num_terms = 10
print_fibonacci(num_terms)
```

```python
#Write a program to find if given number is prime number or not.
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False

    return True
num = 17
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
#Check whether given number is palindrome or not.
def is_palindrome(number):
    str_num = str(number)
    reversed_str = str_num[::-1]
    return str_num == reversed_str

num = 12321
if is_palindrome(num):
    print(f"{num} is a palindrome.")
else:
    print(f"{num} is not a palindrome.")
```

```python
#Write a program to print sum of digits.
def sum_of_digits(number):
    number = abs(number)
    total = 0
    while number > 0:
        total += number % 10
        number //= 10
    return total

num = 12345
print(sum_of_digits(num))
```

```python
#Count and print all numbers divisible by 5 or 7 between 1 to 100.
count = 0
for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1
print(f"\nTotal count: {count}")
```

```python
#Convert all lower cases to upper case in a string.
original_string = "Hello World"
upper_string = original_string.upper()
print(upper_string)
```

```python
#Print the table for a given number: 5 * 1 = 5 5 * 2 = 10………..
number = 5
for i in range(1, 11):
    print(f"{number} * {i} = {number * i}")
```

```python
#Write a program to print the folLowing pattern 123454321 1234 *4321 123 * * 321 12 * * * 21 1 * * * * 1
rows = 5
for i in range(rows, 0, -1):
    for j in range(1, i + 1):
        print(j, end="")
    for j in range(1, 2 * (rows - i)):
        print(" ", end="")
    for j in range(i, 0, -1):
        if j != rows:
            print(j, end="")
    print()
```

```python
#Write a program to print the sum of the following series 1+ ½ + 1/3 + ¼ +….+1/n
def sum_harmonic_series(n):
    total = 0.0
    for i in range(1, n + 1):
        total += 1 / i
    return total

n = 10
result = sum_harmonic_series(n)
print(f"The sum of the series up to 1/{n} is: {result}")
```
